```python
import warnings
import itertools
import pandas as pd
import numpy as np
import statsmodels.api as sm #sarima model
import matplotlib.pyplot as plt
#plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv(path)
df['Month'] = pd.to_datetime(df['Month'])
print(df.head())
```

```python
y = pd.Series(data=df['#Passengers'].values, index=df['Month'])
print(y.head())
```

```python
log_y = np.log(y) #log return
log_y.plot(figsize=(15, 6))
plt.show()
```

```python
# Define the p, d and q parameters to take any value between 0 and 2
#먼저 acf, pacf 등으로 order를 파악해야하지만 많이 해봤으니 바로 정해주자
d = [1]
p = q = range(0, 3) #0~2
```

```python
# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q)) #iteration할 order
```

```python
# Generate all different combinations of seasonal p, q and q triplets
# 얘도 다 원래 정해줘야 하지만 이미 아니까 그냥 넣어줌, seasonal duration 추가
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages
#경고메세지 출력 X
```

```python
best_result = [0, 0, 10000000]
#best_result = [(0, 1, 1), (1, 1, 2, 12), 825.6737919396066]
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = model.fit()
            
            print('ARIMA{} x {} - AIC: {}'.format(param, param_seasonal,
                                                  results.aic))#aic값 구함

            if results.aic < best_result[2]: #원래 값보다 작으면 작은 값으로 바꿔줌
                if sum(param) + sum(param_seasonal[:-1]) <7: # order의 합이 6보다 크면 안되기 때문(마지막 값 period 제외)
                    best_result = [param, param_seasonal, results.aic]
        except:
            continue
print('\nBest Result:', best_result) 
```

```python
best_result = [(0, 1, 1), (1, 1, 2, 12), 825.6737919396066]
model = sm.tsa.statespace.SARIMAX(y,
                                order= best_result[0],
                                seasonal_order=best_result[1],
                                enforce_stationarity=False,
                                enforce_invertibility=False)
```

```python
results = model.fit()
print(results.summary().tables[1])#coefficient 값들을 볼 수 있다(수식 작성 가능)
```

```python
results.plot_diagnostics(figsize=(15, 12))
#plt.show()
```

```python
# Prediction 
pred = results.get_prediction(start=pd.to_datetime('1958-01-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = y['1949':].plot(label='Observed', figsize=(15, 12))
pred.predicted_mean.plot(ax=ax, label='One-Step Ahead Forecast', alpha=.7)

#plotting 옵션
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('1958-01-01'), y.index[-1],
                 alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('Airline Passengers')
plt.legend()
plt.show() #주황색 부분이 예측한 값
```

```python
# Extract the predicted and true values of our time series
y_forecasted = pred.predicted_mean
y_truth = y['1958-01-01':]
```

```python
# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```
